# Análise de Dados PROCON

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando arquivo csv e transforamando em Dataframe
df = pd.read_csv('CNRF_2017.csv', sep= ';')
```

```python
df
```

```python
# Informações gerais do Dataframe
df.info()
```

```python
# Qual a quantidade de chamados por região?
df['regiao'].value_counts()
```

```python
df['regiao'].value_counts().plot.bar(color= 'g')
```

```python
# Qual o estado com maior quantidade de chamados?
Qtd_chamados = df['uf'].value_counts()
Qtd_chamados
```

```python
df['uf'].value_counts().plot.bar()
```

```python
# Número de chamados atendidos por UF?
df_atendidas = df[df['atendida'] == 'S']
```

```python
Qtd_chamdosAtend = df_atendidas['uf'].value_counts()
Qtd_chamdosAtend
```

```python
# Eficiencia de atendimento por UF?
Qtd = (Qtd_chamdosAtend*100)/Qtd_chamados
Qtd
```

```python
Qtd.plot.bar()
```

```python
# Qtd chamdos por sexo?
df['sexoConsumidor'].value_counts()
```

```python
# Qtd chamados atendidos por sexo?
df_atendidas['sexoConsumidor'].value_counts()
```
